# Binary phase diagrams caluation of Hf-Mo-Nb-Ta-Ti-V-W-Zr by pycalphad



**pycalphad** is a free and open source software written in Python for performing thermodynamic calculations within the CALculation of PHAse Diagrams (CALPHAD) method. https://pycalphad.org/docs/latest/#


```python
from pycalphad import Database, binplot, calculate, equilibrium, variables as v
from pycalphad.plot.utils import phase_legend
from eqplot_helper import eqplot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dbf = Database('wzr_guo.tdb') # Load database and choose the phases that will be plotted
comps = ['W', 'ZR', 'VA'] # Set the components to consider, including vacanies (VA) explictly
phases = list(dbf.phases.keys()) # Get all the phases in the database programatically
conds = {v.N: 1.0, v.P: 101325, v.T: (500, 3200, 1), v.X('W'): (0, 1, 0.01)} # Create the dictionary of conditions
binplot(dbf, comps, phases, conds) # Plot by passing in all the variables
```

```python
# Get the colors that map phase names to colors in the legend
legend_handles, color_dict = phase_legend(phases)
fig = plt.figure(figsize=(9,6))
ax = fig.gca()

# Loop over phases, calculate the Gibbs energy, and scatter plot GM vs. X(RE)
for phase_name in phases:
    result = calculate(dbf, ['MO', 'NB'], phase_name, P=101325, T=2800, output='GM')
    ax.scatter(result.X.sel(component='MO'), result.GM, marker='.', s=5, color=color_dict[phase_name])

# Format the plot
ax.set_xlabel('Mole fraction of Mo')
ax.set_ylabel('GM [J/mol-atom]')
ax.set_xlim((0, 1))
ax.legend(handles=legend_handles, loc='center left', bbox_to_anchor=(1, 0.6))
plt.show()
```

```python
#heat cacpactiy
dbf = Database('dft.tdb')
result = calculate(dbf, ['MO'],'BCC_A2', T=(500, 4500, 10), P=101325, N=1, output='CPM')
fig = plt.figure(figsize=(9,6))
fig.gca().set_xlabel('Temperature (K)')
fig.gca().set_ylabel('Isobaric Heat Capacity (J/mol-atom-K)')
fig.gca().plot(result['T'], np.squeeze(result['CPM']))
plt.legend('MO')
plt.show()
```
